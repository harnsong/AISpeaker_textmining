````python
import os
import sys
import urllib.request
import requests
import json
import math
````

````python
NAVER_URL = "https://openapi.naver.com/v1/search/news.json"

API_PARAMETER = {
    "client_id": "E5cL2__bFGNST8f4eknt",
    "client_secret": "kcwtUqHT98"
}

QUERY_PARAMETER = {
    "query": "네이버 클로바 스피커",
    "display": 100,
    "start": 1,
    "sort": "sim"
}
````

````python
def naver_api_blog_search(api_parameter, query_parameter):
    result = None
    
    client_id = api_parameter['client_id']
    client_secret = api_parameter['client_secret']
    
    response = requests.get(NAVER_URL, headers={
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret
    }, params=query_parameter)
    
    rescode = response.status_code
    if(rescode==200):
        result = response.text
    else:
        print("Error Code:" + str(rescode))
    return result

def get_search_result_page_count(api_parameter, query_parameter):
    search_result_page_count = 0
    display_count = query_parameter['display']
    
    decoded_response_body = naver_api_blog_search(api_parameter, query_parameter)
    if decoded_response_body:
        json_response_body = json.loads(decoded_response_body)
        
        total_posts = json_response_body['total']
        if total_posts != 0:
            search_result_page_count = math.ceil(total_posts / display_count)
    return search_result_page_count

def get_urls(api_parameter, query_parameter, search_result_page_count):
    display_count = query_parameter['display']
    url_list = []
    for i in range(search_result_page_count):
        start = display_count*i + 1
        if start > 1000:
            start = 1000
        
        query_parameter['start'] = start
        decoded_response_body = naver_api_blog_search(api_parameter, query_parameter)
        if decoded_response_body:
            json_response_body = json.loads(decoded_response_body)
            
            for dp in json_response_body['items']:
                url_list.append(dp['link'].replace("amp;", ""))
                print(dp['title'])
        
        ### 테스트를 위해 넣은 부분 ###
        #if i == 1:
        #    break
        #########################
        
        if start == 1000:
            break
    return url_list


if __name__ == '__main__':
    api_parameter = API_PARAMETER
    query_parameter = QUERY_PARAMETER
    
    p_count = get_search_result_page_count(api_parameter, query_parameter)
    urls = get_urls(api_parameter, query_parameter, p_count)
````

````python
import requests
from bs4 import BeautifulSoup

import time
import datetime
import re
````

````python
def clear_content(text):
    # <Reference>
    # https://github.com/lumyjuwon/KoreaNewsCrawler/blob/master/newscrawler/articleparser.py
    
    info_symbol = re.compile('[\(]') # 이부분은 새로 추가함 ['커머스(전자상거래)와' -> '커머스전자상거래와' 이렇게 변하던것을 -> '커머스 전자상거래' ]
    special_symbol = re.compile('[\{\}\[\]\/?,;:|\)*~`!^\-_+<>@\#$%&n▲▶◆◀■\\\=\(\'\"]')
    content_pattern = re.compile('본문 내용|TV플레이어| 동영상 뉴스|flash 오류를 우회하기 위한 함수 추가fuctio flashremoveCallback|tt|t|앵커 멘트|xa0')
    
    # 기사 본문에서 필요없는 특수문자 및 본문 양식 등을 다 지움
    info_symbol_removed_content = re.sub(info_symbol, ' ', text) # 이부분은 새로 추가함 ['커머스(전자상거래)와' -> '커머스전자상거래와' 이렇게 변하던것을 -> '커머스 전자상거래' ]
    special_symbol_removed_content = re.sub(special_symbol, '', info_symbol_removed_content)
    end_phrase_removed_content = re.sub(content_pattern, '', special_symbol_removed_content)
    blank_removed_content = end_phrase_removed_content.strip().replace('   ', '')  # 공백 에러 삭제
    reversed_content = ''.join(reversed(blank_removed_content))  # 기사 내용을 reverse 한다.
    content = ''
    for i in range(0, len(blank_removed_content)):
        # reverse 된 기사 내용중, ".다"로 끝나는 경우 기사 내용이 끝난 것이기 때문에 기사 내용이 끝난 후의 광고, 기자 등의 정보는 다 지움
        if reversed_content[i:i + 2] == '.다':
            content = ''.join(reversed(reversed_content[i:]))
            break
    return content

"""
def clear_headline(text):
    # 기사 제목에서 필요없는 특수문자들을 지움
    special_symbol = re.compile('[\{\}\[\]\/?,;:|\)*~`!^\-_+<>@\#$%&n▲▶◆◀■\\\=\(\'\"]')
    
    special_symbol_removed_headline = re.sub(special_symbol, '', text)
    return special_symbol_removed_headline
"""
````

````python
if __name__ == '__main__':
    
    company_list = []
    section_list = []
    title_list = []
    publish_datetime_list = []
    final_datetime_list = []
    texts_list = []
    
    for i, url in enumerate(naver_news_urls):    

        # 실제 데이터가 들어있는 url 추출
        r = requests.get(url)
        soup = BeautifulSoup(r.text.encode("utf-8"), "html.parser")
        
        company = get_publishing_company_navernews(soup)
        company_list.append(company)
        
        section = get_section_navernews(soup)
        section_list.append(section)
        
        # 기사 제목 추출
        title = get_title_navernews(soup)
        title_list.append(title)

        # 기사 입력일 추출
        publish_datetime = get_publish_datetime_navernews(soup)
        publish_datetime_list.append(publish_datetime)
        
        final_datetime = get_final_datetime_navernews(soup)
        final_datetime_list.append(final_datetime)
        
        # 기사 텍스트 추출
        texts = get_texts_navernews(soup)
        texts_list.append(texts)
        
        print(i+1, "finished")

        time.sleep(0.01)     
````
