### 'AI 스피커'
* 4506개

```python
list_of_tweets = query_tweets("AI 스피커", 
                              begindate=datetime.date(2016,10,1), 
                              enddate=datetime.date(2019,5,20), 
                              #poolsize=150,
                              lang='ko')
for tweet in list_of_tweets:
    print(tweet.fullname, tweet.timestamp, tweet.text)
    print("\n")
    
    #트윗 내용 외에도 아이디, 멘션, 계정 아이디 등에 키워드가 있으면 긁어옴
```

```python
len(list_of_tweets)
```

```python
def process_results(list_of_tweets):
    id_list = [tweet.id for tweet in list_of_tweets]
    data_set = pd.DataFrame(id_list, columns=["id"])
    
    #Processing Tweet Data
    data_set["text"] = [tweet.text for tweet in list_of_tweets]
    data_set["created_at"] = [tweet.timestamp for tweet in list_of_tweets]
    data_set["retweet_count"] = [tweet.retweets for tweet in list_of_tweets]
    data_set["like_count"] = [tweet.likes for tweet in list_of_tweets]
    #data_set["source"] = [tweet.source for tweet in list_of_tweets]
    data_set["reply_count"] = [tweet.replies for tweet in list_of_tweets]
    
    #Processing User Data
    data_set["user_name"] = [tweet.fullname for tweet in list_of_tweets]
    data_set["user_id"] = [tweet.user for tweet in list_of_tweets]
    #data_set["user_screen_name"] = [tweet.author.screen_name for tweet in list_of_tweets]
    #data_set["user_name"] = [tweet.author.name for tweet in list_of_tweets]
    #data_set["user_created_at"] = [tweet.author.created_at for tweet in list_of_tweets]
    #data_set["user_description"] = [tweet.author.description for tweet in list_of_tweets]
    #data_set["user_followers_count"] = [tweet.author.followers_count for tweet in list_of_tweets]
    #data_set["user_friends_count"] = [tweet.author.friends_count for tweet in list_of_tweets]
    #data_set["user_location"] = [tweet.author.location for tweet in list_of_tweets]
    
    return data_set
```

```python
df_1 = process_results(list_of_tweets)
```

```python
df_1.tail()
```

```python
df_1.to_csv("twitterscraper_ai스피커_0520.csv")
```
